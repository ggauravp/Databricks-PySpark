```python
df = spark.table("workspace.default.movies")
df.display()
```

```python
df.show(4)
```

```python
display(df.describe())
```

## column filtering

```python
df_colm_filter = df.select("title", "budget", "imdb_rating")
df_colm_filter.show(4, truncate=False)
```

## Row filtering

```python
df_row_filter = df.filter(df.release_year > 2000)
df_row_filter.show(4)
```

```python
from pyspark.sql.functions import col
df_row_filter1 = df.filter(col("release_year")>=2000)  
df_row_filter1.show(4)
```

```python
unique_industries = df.select("industry").distinct()
unique_industries.show()
```

```python
df = df.withColumn("profit", col("revenue") - col("budget"))
df.display()
```
