## Creating DataFrame using List

```python
data = [
    ("2017-01-01", 32.0,  6.0,  "Rain"),
    ("2017-01-04", None,  9.0,  "Sunny"),
    ("2017-01-05", 28.0,  None, "Snow"),
    ("2017-01-06", None,  7.0,  None),
    ("2017-01-07", 32.0,  None, "Rain"),
    ("2017-01-08", None,  None, "Sunny"),
    ("2017-01-09", None,  None, None),
    ("2017-01-10", 34.1,  8.1,  "Cloudy"),
    ("2017-01-11", 40.0, 12.0,  "Sunny"),
]
schema = "day string, temp double, windspeed double, event string"
df = spark.createDataFrame(data, schema)

df.show()
df.printSchema()
```

```python
from pyspark.sql import functions as f, types as T
df = df.withColumn("day", f.to_date("day", "yyyy-MM-dd"))  # normalize to DateType
df.printSchema()
```

## Converting DataFrame into Sql table [VIEW]

```python
df.createOrReplaceTempView("weather") # create a temp view for SQL i.e for this session only
```

```python
spark.sql("select * from weather").show()
```

```python
%sql
select * from weather
```
